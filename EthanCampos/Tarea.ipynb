### Tiro Vertical ###

Sabemos que: 

$$
\textbf{V} = at + Vo
$$

$$
\textbf{y} = \left(-\frac{1}{2}gt^2+v_{0}t+y_0\right)
$$

$$
\textbf{a} = -g$$

Poniendo los valores:

```julia
V0 = 10;
```

```julia
g = 9.81;
```

```julia
a = -g;
```

```julia
t = collect(0:1.0:20);
```

Igualamos y=0

$$
\textbf{y} = \left(-\frac{1}{2}gt^2+v_{0}t\right)\
$$

```julia
Y = zeros(t);
```

```julia
for i ∈ eachindex(t)
    Y[i] = (-(1/2*g*t[i]^2)+V0*t[i])
end
```

```julia
using Plots
```

```julia
pyplot();
```

```julia
plot(Y,w=2)
title!("Tiro Vertical")
xaxis!("Tiempo")
yaxis!("Posición")
```

```julia
V = zeros(t);
```

```julia
for i ∈ eachindex(t)
    V[i] = a*t[i]+V0
end
```

```julia
plot(t,V,w=2)
title!("Tiro vertical")
xaxis!("Tiempo")
yaxis!("Velocidad")
```

### Tiro Parabólico ###

Sabemos que para calcular el Alcance Máximo usamos la siguiente formula:

$$
{x_{máx}} = {v_{0}}{cos(θ)}{t}
$$

Y para el Tiempo de Vuelo: 
$$
                y = -\frac{1}{2}gt^2+v_{0y}{sin(θ)}{t}
                $$


Definiendo Valores:

```julia
V0 = 100.0;
```

```julia
θ = π/4;
```

```julia
g = 9.81;
```

```julia
t = collect(0:1.0:15.0);
```

```julia
x = zeros(t);
Y = zeros(t);
```

```julia
for i ∈ eachindex(t)
    x[i] =  x[i] = V0*cos(θ)*t[i]
    Y[i] = (-(1/2*g*t[i]^2)+V0*sin(θ)*t[i])
end
```

```julia
plot(t,x,w=2)
title!("Alcance Máximo")
xaxis!("Tiempo")
yaxis!("Distancia")
```

```julia
plot(t,Y,w=2)
title!("Tiempo de Vuelo")
xaxis!("Tiempo Transcurrido")
yaxis!("Altura")
```
